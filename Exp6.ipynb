```python
#importing the basic library
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
print(x)
```

```python
y=2*x
```

```python
y
```

line chart

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

bar chart

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
a=(1,3,6,7,8,9,4,6)
b=(11,22,33,45,67,77,98,76)
plt.scatter(a,b)
plt.title("scatter plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.scatter(a,b,color='brown',edgecolor='yellow')
plt.show()
```
